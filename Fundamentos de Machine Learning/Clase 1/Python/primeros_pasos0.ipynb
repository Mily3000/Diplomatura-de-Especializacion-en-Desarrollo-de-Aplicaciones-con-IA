### 4.- Presente un análisis exploratorio del nivel educacional (“nivel_educt”), y edad (“edad”), utilice los estadíticos y gráficos más adecuados para cada caso.

```
# Analisis Nivel Educacional
df.nivel_educt.describe()
    
tabla_freq = df.nivel_educt.value_counts()  
    
#Moda
moda = tabla_freq.sort_values(ascending=False).index[0]
    
#Tasa de variación
fm = tabla_freq.sort_values(ascending=False)[0]/tabla_freq.sum()
V = 1 - fm  
```

```
# Graph 1
tabla_freq.plot(kind="pie",
                label="",
                legend=True, 
                autopct='%.2f%%', 
                labels=None, 
                figsize=(6,6), fontsize=15)
```

```
# Graph 2
fig, ax = plt.subplots()
fig.set_size_inches(6,6)
wedges, texts, autotexts = ax.pie(x=tabla_freq,autopct='%.2f%%')
leg = ax.legend(wedges, tabla_freq.index.values,
                bbox_to_anchor=(0.9, 0.9),
                fontsize=12)
leg.set_title("Educación", prop = {'size':12,'weight':'bold'})
plt.setp(autotexts, fontsize=12, fontweight="bold")

ax.set_title("Nivel Educacional",fontdict={'fontsize':14, 'fontweight':'bold'})
plt.show()
```

```
# Graph 3
sns.set(style="whitegrid")
fig, ax = plt.subplots()
fig.set_size_inches(7,5)
ax = sns.barplot(y = tabla_freq.index.values, 
                 x = tabla_freq.values) 
#ax.tick_params(axis="y",labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.set_title("Nivel Educacional",
             fontdict={'fontsize':14, 'fontweight':'bold'})
j = 0.1
for xpos in tabla_freq.values:
  xporc = str(round(100*xpos/tabla_freq.values.sum(),2))+" %"
  ax.text(xpos,j,xporc)
  j+=1
plt.show()
```

### 6.- Obtenga la Desviación Estándar, Desviación Media y MEDA del ingreso familiar. Explique que buscan estos estadisticos y comente a que puede deberse la diferencia (en base a los datos) entre los tres indicadores.

```
print(df.ingres_famim.std())
    
print((abs(df.ingres_famim - df.ingres_famim.mean())).mean())

print((abs(df.ingres_famim - df.ingres_famim.median())).median())
```

### 7- Si en ambos casos se descartan los outlier, ¿como cambian los estadisticos anteriores?. Comente observaciones.

```
from matplotlib.cbook import boxplot_stats
outliers = boxplot_stats(df.ingres_famim)[0]['fliers']
   
indices = df.ingres_famim < outliers.min()
   
df.ingres_famim[indices].mean()
df.ingres_famim[indices].median()
df.ingres_famim[indices].value_counts().sort_values(ascending=False).index[0]
   
   
df.ingres_famim[indices].std()
(abs(df.ingres_famim[indices] - df.ingres_famim[indices].mean())).mean()
(abs(df.ingres_famim[indices] - df.ingres_famim[indices].median())).median()
   
```
