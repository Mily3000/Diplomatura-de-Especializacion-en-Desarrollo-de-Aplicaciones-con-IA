# Join en Python 

```
cnx = mysql.connector.connect(user='fkAgdxoNxs', password='GbjW33NDFc', host='remotemysql.com', database='fkAgdxoNxs')
cursor = cnx.cursor()
cursor.execute("select * from Customers;")
result = cursor.fetchall()
  
customers = pd.DataFrame(result)

cursor = cnx.cursor()
cursor.execute("SELECT COLUMN_NAME FROM INFORMATION_SCHEMA.COLUMNS WHERE TABLE_NAME = 'Customers' ORDER BY ORDINAL_POSITION ASC")
result = cursor.fetchall()

cnx.close()
  
customers.columns = pd.DataFrame(result)[0]
```

```
customers.head(4)
```

# Primeras visualizaciones

```
cnx = mysql.connector.connect(user='fkAgdxoNxs', password='GbjW33NDFc', host='remotemysql.com', database='fkAgdxoNxs')
cursor = cnx.cursor()
cursor.execute("SELECT OrderID, SUM(Quantity*UnitPrice) TOTAL FROM `Order Details` GROUP BY OrderID")
montos = cursor.fetchall()
montos = pd.DataFrame(montos)
cnx.close()

montos.columns = ['OrderID','TOTAL']

fulldata = pd.merge(fulldata,montos,on='OrderID')
fulldata.head(3)
```

```
import seaborn as sns

sns.scatterplot(x="OrderID", y="TOTAL", data=fulldata)
```

```
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
sns.set(style="whitegrid")
sns.scatterplot(x="OrderID", y="TOTAL", data=fulldata,)
```

```
plt.figure(figsize=(10,5))
sns.set(style="whitegrid")
sns.scatterplot(x="OrderID", y="TOTAL", data=fulldata,)
plt.title('TOTAL vs OrderID')
plt.xlabel('ID')
plt.ylabel('Total $')
```
