# Recordando SVM

Empecemos retomando la clasificación del conjunto de datos 2 del ejercicio anterior con Support Vector Machines. Por ser un conjunto no separable linealmente, usamos un kernel gaussiano (RBF). Para efectos de comparación posterior, no normalizaremos el conjunto de datos.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#import seaborn as sns; sns.set()
%matplotlib inline
from sklearn.svm import SVC # "Support vector classifier"

# Función de ayuda para visualización de frontera de decisión
def plot_svc_decision_function(model, ax=None, plot_support=True, levels=[-1, 0, 1], linestyles=['--', '-', '--']):
    # Tomado del libro 'Python Data Science Handbook' de Jake VanderPlas
    """Plot the decision function for a 2D SVC"""
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    
    # plot decision boundary and margins
    ax.contour(X, Y, P, colors='k',
               levels=levels, alpha=0.5,
               linestyles=linestyles)
    
    # plot support vectors
    if plot_support:
        ax.scatter(model.support_vectors_[:, 0],
                   model.support_vectors_[:, 1],
                   s=300, linewidth=1, facecolors='none');
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)

    
# Carga de datos
data2 = pd.read_csv('ex6data2.csv', names=['x1', 'x2', 'y'])
X = data2[['x1', 'x2']]
y = data2['y']

# Ajuste de modelo con kernel gaussiano
model = SVC(kernel='rbf', C=1, gamma=50)
model.fit(X, y)

# Visualización
plt.title('Frontera de decision SVM con kernel gaussiano (Conjunto de datos 2)')
plt.margins(x=0, y=0)
plt.xlabel('x1')
plt.ylabel('x2')
plt.scatter(X['x1'], X['x2'], c=y, s=10, cmap='prism');
plot_svc_decision_function(model, plot_support=False, levels=[0], linestyles=['-']) 
plt.show()

# Evaluación
score = model.score(X, y)
tasa_de_error = (1- score)
n_ejemplos = len(X)

print ('Total de ejemplos: %d' % n_ejemplos)
print ('Exactitud: %0.4f' % score)
print ('Tasa de error: %0.4f' % tasa_de_error)
```

# Entrenamiento de un árbol de decisión con CART

En este ejercicio usaremos la implementación de árboles de decisión de `scikit-learn`, la cual emplea una versión optimizada de CART. Ver la documentación oficial en [1.10. Decision Trees](http://scikit-learn.org/stable/modules/tree.html)

## Primera partición

El ajuste del modelo es muy sencillo. Limitemos la profundidad `max_depth=1` para visualizar sólo una primera partición. Establecemos el parámetro `random_state=0` sólo para asegurar resultados reproducibles. En la práctica se aconseja probar con diferentes valores y conservar el que mejor resultados produzca.

```python
from sklearn.tree import DecisionTreeClassifier

# Ajuste de modelo con árbol de decisión
treemodel = DecisionTreeClassifier(max_depth=1, random_state=0)
treemodel.fit(X, y)
```

### Función de ayuda

Adaptemos nuestra función de visualización. Cambiamos la función `decision_function` por la función `predict` y aprovechamos para eliminar los parámetros propios de SVC que no aplican para los árboles de decisión.

```python
# Función de ayuda para visualización de frontera de decisión
def plot_tree_decision_function(model, ax=None):
    # Adaptado del libro 'Python Data Science Handbook' de Jake VanderPlas
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1], 300)
    y = np.linspace(ylim[0], ylim[1], 300)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.predict(xy).reshape(X.shape)
    
    # plot decision boundary and margins
    ax.contour(X, Y, P, colors='k',
               levels=[0], alpha=0.5,
               linestyles=['-'])
    
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
```

### Evaluación

Examina la primera división realizada por el árbol de decisión.

```python
# Visualización
plt.title('Frontera de decision con max_depth=1 (Conjunto de datos 2)')
plt.margins(x=0, y=0)
plt.xlabel('x1')
plt.ylabel('x2')
plt.scatter(X['x1'], X['x2'], c=y, s=10, cmap='prism');

plot_tree_decision_function(treemodel) 
plt.show()

# Evaluación
score = treemodel.score(X, y)
tasa_de_error = (1 - score)
n_ejemplos = len(X)

print ('Total de ejemplos: %d' % n_ejemplos)
print ('Exactitud: %0.4f' % score)
print ('Tasa de error: %0.4f' % tasa_de_error)
```

## Optimización de los parámetros

Para esta parte final del ejercicio, usaremos el conjunto de datos 3.

```python
data3train = pd.read_csv('ex6data3train.csv', names=['x1', 'x2', 'y'])
X = data3train[['x1', 'x2']]
y = data3train['y']

data3val = pd.read_csv('ex6data3val.csv', names=['x1', 'x2', 'y'])
Xval = data3val[['x1', 'x2']]
yval = data3val['y']

plt.title('Conjunto de datos 3')
plt.margins(x=0, y=0)
plt.xlabel('x1')
plt.ylabel('x2')
plt.scatter(X['x1'], X['x2'], c=y, s=10, cmap='prism', label='Conjunto de entrenamiento')
plt.scatter(Xval['x1'], Xval['x2'], c=yval, marker='x', s=20, cmap='prism', label='Conjunto de validacion')
plt.show()
```

Usando la técnica aprendida en la clase anterior, haz una búsqueda de la mejor combinación de valores para los parámetros `max_depth` y `min_samples_leaf`. 

Completa el código en los dos puntos indicados. Emplea `random_state=0` para que los resultados sean reproducibles.

```python
mejor_modelo = None
mejor_score = 0
for max_depth in range(1, 11):
    for min_samples_leaf in range(1, 11):

        model = DecisionTreeClassifier(max_depth=max_depth, 
                                       min_samples_leaf=min_samples_leaf, 
                                       random_state=0)  ## COMPLETAR...
        model.fit(X, y)

        score_val = model.score(Xval, yval) ## COMPLETAR...

        if score_val > mejor_score:
            mejor_score = score_val
            mejor_modelo = model

            
# Visualización
plt.title('Arbol de decisión (Conjunto de datos 3)')
plt.margins(x=0, y=0)
plt.xlabel('x1')
plt.ylabel('x2')
plt.scatter(X['x1'], X['x2'], c=y, s=10, cmap='prism');
plt.scatter(Xval['x1'], Xval['x2'], c=yval, marker='x', s=20, cmap='prism', label='Conjunto de validacion')
plot_tree_decision_function(model) 
plt.show()

# Evaluación
print ('Mejor valor de max_depth: %d' % mejor_modelo.get_params()['max_depth'])
print ('Mejor valor de min_samples_leaf: %d' % mejor_modelo.get_params()['min_samples_leaf'])
print ('Exactitud en el conjunto de entrenamiento: %0.4f' % mejor_modelo.score(X, y))
print ('Exactitud en el conjunto de validación: %0.4f' % mejor_modelo.score(Xval, yval))
print ('Tasa de error en el conjunto de validación: %0.4f' % (1 - mejor_modelo.score(Xval, yval)))
```
